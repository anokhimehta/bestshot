```python
import os
import swiftclient
from dotenv import load_dotenv
```

```python
load_dotenv('.env')

conn = swiftclient.Connection(
    auth_version='3',
    authurl=os.environ['OS_AUTH_URL'],
    os_options={
        'application_credential_id': os.environ['OS_APPLICATION_CREDENTIAL_ID'],
        'application_credential_secret': os.environ['OS_APPLICATION_CREDENTIAL_SECRET'],
        'region_name': os.environ['OS_REGION_NAME'],
        'auth_type': 'v3applicationcredential'
    }
)

# Test: list top-level contents of the bucket
_, objects = conn.get_container('ak12754-data-proj19')
for obj in objects[:10]:
    print(obj['name'])
```

```python
# Check CSV exists at expected path
_, content = conn.get_object('ak12754-data-proj19', 'koniq10k/koniq10k_scores_and_distributions.csv')
import pandas as pd
import io
df = pd.read_csv(io.BytesIO(content))
print(f"CSV shape: {df.shape}")
print(df.columns.tolist())
print(df.head(3))
```
